Dropping "Unnamed: 1" column from our dataset as it is not useful

```python
data.drop("Unnamed: 1", axis=1, inplace=True)
data.info()
```

DATA PRE-PROCESSING

```python
data.columns = data.columns.str.replace(' ','')
data.columns = data.columns.str.replace('_','.')
data.columns
```

```python
data.isna().sum()
```

```python
data.dropna(axis=0,subset=['Location'],inplace=True)
```

Label Encoding

```python
col_list = []
for i in data.columns:
    if((data[i].dtypes=='object') and (i!='JobTitle')):
        col_list.append(i)

col_list
```

```python
from sklearn.preprocessing import LabelEncoder

labelencoder = LabelEncoder()

for i in col_list:
    data[i]=labelencoder.fit_transform(data[i])

data

```

dropping "Longitude", "Latitude", "Sal" from the dataset

```python
data = data.drop(['Longitude'],axis=1)
data = data.drop(['Latitude'],axis=1)
data = data.drop(['sal'],axis=1)
```

```python
data
```

Feature attributes is X and Target attribute is Y

```python
data
```

```python
x = data.drop(['JobTitle'], axis = 1)
print(x)
y = data.JobTitle
```

```python
print(y)
```

Splitting into Training and Testing sets

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.20,random_state=42)

print("Training dataset dimension: ",x_train.shape)
print("Testing dataset dimension: ",x_test.shape)
```

Building our model

```python
from sklearn.linear_model import LogisticRegression

logmodel = LogisticRegression()

logmodel.fit(x_train,y_train)
```

Prediction using x_test

```python
prediction = logmodel.predict(x_test)
```

Reports for comparision of actual output(y_test) and our predicted output(prediction)

```python
from sklearn.metrics import classification_report

print(classification_report(prediction, y_test))
```
